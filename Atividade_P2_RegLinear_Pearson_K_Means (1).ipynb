#Análise de PIB com K-Means (Crescimento do PIB (2012-2022))


Esta análise teve como objetivo segmentar os países em grupos (clusters) com base na sua dinâmica de crescimento económico, em vez da sua riqueza absoluta. Para isso, utilizámos a taxa de crescimento do PIB per capita entre 2012 e 2022 como a métrica principal para a clusterização K-Means.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import numpy as np
```

##1. Carregar os dados


```python
df_pib = pd.read_csv('pib.csv')

print("Primeiras 5 linhas do DataFrame:")
df_pib.head()
```

##2. Pré-processamento e Cálculo do Crescimento

```python
# Removendo colunas que não vamos utilizar na análise
df_pib = df_pib.drop(['Country Code','Indicator Name', 'Indicator Code'], axis=1)

display(df_pib.head())
```

```python
anoInicio = '2012'
anoFim = '2022'

# Criar um novo DataFrame para a análise

df_crescimento = df_pib[['Country Name', anoInicio, anoFim]].copy()
df_crescimento.head()
```

```python
# Limpeza das colunas de PIB para os dois anos
for ano in [anoInicio, anoFim]:
    df_crescimento[ano] = df_crescimento[ano].astype(str).str.replace(',','.', regex=False)
    df_crescimento[ano] = pd.to_numeric(df_crescimento[ano], errors='coerce')

df_crescimento.sample(10)
```

```python
#Remover linhas com dados vazios
df_crescimento = df_crescimento.dropna()
df_crescimento
```

```python
# Calcular a taxa de crescimento percentual
# Fórmula: ((Valor Final - Valor Inicial) / Valor Inicial) * 100
df_crescimento['Crescimento %'] = ((df_crescimento[anoFim] - df_crescimento[anoInicio]) / df_crescimento[anoInicio]) * 100
df_crescimento.head()
```

```python
# Tratar casos de divisão por zero ou valores infinitos que podem surgir
df_crescimento.replace([np.inf, -np.inf], np.nan, inplace=True)
df_crescimento.dropna(subset=['Crescimento %'], inplace=True)
df_crescimento.head()
```

```python
print(f"Número de países com dados de crescimento válidos (2012-2022): {len(df_crescimento)}")
```

##3. Clusterização com K-Means

```python
# Selecionar a feature(X) (Crescimento_%) e padronizar os dados
X = df_crescimento[['Crescimento %']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Método do Cotovelo (Elbow Method) para encontrar o K ideal
wcss = []
k_range = range(1, 11)
for i in k_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Gerar o gráfico do Método do Cotovelo
plt.figure(figsize=(10, 5))
plt.plot(k_range, wcss, marker='o', linestyle='--')
plt.title('Método do Cotovelo para Clusters de Crescimento')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('WCSS')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

A partir do ponto 4 vemos que não há um ajuste tão brusco

```python
# O "cotovelo" sugere que K=4 é uma boa escolha para detalhar os grupos de crescimento
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
df_crescimento['Cluster'] = kmeans.fit_predict(X_scaled)
df_crescimento.head()
```

##4. Análise e Visualização dos Resultados

```python
# Obter os centros dos clusters (taxa de crescimento média de cada grupo)
cluster_centers_inversed = scaler.inverse_transform(kmeans.cluster_centers_)

# Ordenar os clusters pela taxa de crescimento
sorted_clusters_indices = sorted(range(n_clusters), key=lambda k: cluster_centers_inversed[k][0])
cluster_mapping = {old_label: new_label for new_label, old_label in enumerate(sorted_clusters_indices)}
df_crescimento['Cluster'] = df_crescimento['Cluster'].map(cluster_mapping)
cluster_centers_inversed = sorted(cluster_centers_inversed, key=lambda x: x[0])
```

```python
# Imprimir a análise dos clusters
print("\n--- Análise dos Clusters por Crescimento do PIB (2012-2022) ---")
growth_labels = ["Estagnação/Recessão", "Baixo Crescimento", "Crescimento Moderado", "Alto Crescimento"]
for i in range(n_clusters):
    cluster_data = df_crescimento[df_crescimento['Cluster'] == i]
    print(f"\n--- Cluster {i}: {growth_labels[i]} (Crescimento médio: {cluster_centers_inversed[i][0]:,.2f}%) ---")
    print(f"Número de países: {len(cluster_data)}")
    print("Países neste cluster (Top 5 por crescimento):")
    print(cluster_data.sort_values('Crescimento %', ascending=False).head(5)['Country Name'].tolist())
```

```python
# Boxplot para visualizar a distribuição do crescimento
plt.figure(figsize=(12, 8))
sns.boxplot(x='Cluster', y='Crescimento %', data=df_crescimento, palette='plasma')
plt.title('Distribuição do Crescimento do PIB (2012-2022) por Cluster', fontsize=16)
plt.xlabel('Cluster de Crescimento')
plt.ylabel('Crescimento do PIB per capita (%)')
plt.gca().set_xticklabels(growth_labels)
plt.show()
```

```python
# Gráfico de Dispersão
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Country Name', y='Crescimento %', hue='Cluster', data=df_crescimento,
                palette='plasma', s=100, legend='full')
plt.title('Clusters de Países por Crescimento do PIB (2012-2022)', fontsize=16)
plt.xlabel('Países (não em escala)')
plt.ylabel('Crescimento do PIB per capita (%)')
plt.xticks([])
plt.grid(True, which="both", ls="--")
plt.legend(title='Cluster de Crescimento', labels=growth_labels)
plt.show()
```

 É notável a grande distância vertical entre os clusters, especialmente o de "Alto Crescimento" em comparação com os outros. O grupo de "Estagnação/Recessão" agrupa-se na parte inferior, com valores negativos. A maioria dos países concentra-se nos clusters de "Baixo" e "Moderado Crescimento", formando a maior parte da economia global em termos de número de nações
